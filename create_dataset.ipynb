```python
import os
import shutil
import pandas as pd
import natsort
```

```python
path = '/data/data breast cancer/hasil_docking_breast_cancer'
```

```python
directory = natsort.natsorted(os.listdir(path))
```

```python
directory
```

```python
hasil = []
target = []
cid=[]
for j in range(len(directory)):
    
    file = natsort.natsorted(os.listdir(path + '/' + directory[j]))
    for i in range (len(file)):
        data = pd.read_csv(path + '/' + directory[j] + '/' + file[i], header = None)
        hasil.append(data)
        protein = directory[j].split('_')
        target.append(protein[0])
        obat = file[i].split('_')
        cid.append(obat[0])
        
    
    
    
```

```python
hasil_split=[]
for i in range(len(hasil)):
    split = hasil[i]
    split = split[0].str.split('|')
    hasil_split.append(split[0][1])
```

```python
target
```

```python
cid
```

```python
final = pd.DataFrame(list(zip(cid,target,hasil_split)),columns = ['cid','pdb_id','score'])
```

```python
final
```

```python
import pubchempy as pcp
from pubchempy import get_compounds, Compound
```

```python
list_smiles = []
for i in final['cid']:
    canon_smiles = pcp.Compound.from_cid(i)
    print(canon_smiles.canonical_smiles)
    list_smiles.append(canon_smiles.canonical_smiles)
```

```python
final.insert(1, 'smiles', list_smiles) 
```

```python
virus = []
for i in range(len(final)):
    virus.append('breast_cancer')
```

```python
final['fungsi_obat'] = virus
```

```python
data_pubchem = pd.read_csv('/data/data breast cancer/breast_cancer.csv')
```

```python
nama_obat = []
for i in range(len(final)):
    for j in range(len(data_pubchem)):
        print('final = ' + final['cid'][i])
        print(data_pubchem['cid'][j])
        if str(final['cid'][i]) == str(data_pubchem['cid'][j]):
            nama_obat.append(data_pubchem['cmpdname'][j])
            print('nama = ' + data_pubchem['cmpdname'][j])
        
```
