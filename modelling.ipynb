```python
import os
import numpy as np
import pandas as pd

import tempfile

from rdkit import Chem
from rdkit.Chem import AllChem
import deepchem as dc
import tensorflow as tf

from deepchem.utils import download_url, load_from_disk
```

```python
data = pd.read_csv('/data/database/dataset_semua.csv')
```

```python
data = pd.DataFrame(data)
```

```python
data
```

```python
protein = data['pdb_id']
```

```python
ligands = data['smiles']
```

```python
featurizer_1 = dc.feat.MolGraphConvFeaturizer(use_edges=True)
featurizer_2 = dc.feat.CircularFingerprint(size=2048, radius=8)
```

```python
out_1 = featurizer_1.featurize(ligands)
out_2 = featurizer_2.featurize(ligands)
```

```python
final_feature = pd.DataFrame(out_2)
```

```python
final_feature
```

```python
labels = data['score']
```

```python
protein_id = pd.DataFrame(protein).rename(columns = {0 : 'pdb_id'})
```

```python
protein_id = pd.get_dummies(protein_id, columns = ['pdb_id'])
```

```python
protein_id
```

```python
all_feature = pd.concat([final_feature, protein_id], axis = 1)
```

```python
all_feature
```

```python
dataset_model = dc.data.NumpyDataset(X=all_feature, y=labels)
train_dataset, test_dataset = dc.splits.RandomSplitter().train_test_split(dataset_model, seed=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

from deepchem.utils.evaluate import Evaluator
import pandas as pd
```

```python
#RANDOM FOREST MODEL
```

```python
%time
seed = 42 # Set a random seed to get stable results
sklearn_model = RandomForestRegressor(n_estimators=100, max_features='sqrt')
sklearn_model.random_state = seed
model = dc.models.SklearnModel(sklearn_model)
model.fit(train_dataset)
```

```python
# use Pearson correlation so metrics are > 0
metric_pearson = dc.metrics.Metric(dc.metrics.pearson_r2_score)
metric_mae = dc.metrics.Metric(dc.metrics.mae_score)

evaluator = Evaluator(model, train_dataset, [])
train_mae_score = model.evaluate(train_dataset, [metric_mae])
print("RF Train set mae %f" % (train_mae_score["mae_score"]))

evaluator = Evaluator(model, test_dataset, [])
test_mae_score = model.evaluate(test_dataset, [metric_mae])
print("RF Test set R^2 %f" % (test_mae_score["mae_score"]))
```

```python
# Compare predicted and true values
list(zip(model.predict(train_dataset), train_dataset.y))[:5]
```

```python
prediksi_seluruh = model.predict(dataset_model)
```

```python
dataset_RF = data
```

```python
dataset_RF.insert(5, 'prediksi', prediksi_seluruh)
```

```python
dataset_RF
```

```python
dataset_RF.to_csv('data/database/prediksi_dataset_semua.csv',index=False)
```

```python
print('MAE = ' + str(sum(selisih)/len(dataset_RF)))
```
