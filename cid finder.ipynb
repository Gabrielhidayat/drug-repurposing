```python
import pubchempy
from pubchempy import get_compounds, Compound
import pandas as pd
```

```python
database = pd.read_csv('/drugs/database_final.csv')
```

```python
df = pd.read_csv('/drugs/antibiotics/antibiotics.csv', header = None)
df = pd.DataFrame(df[0])
df = pd.DataFrame(df).rename(columns  = { 0 : 'nama_obat'})
```

```python
df
```

```python
nama_virus = []
for i in df['nama_obat']:
    nama_virus.append('antibiotics')
df['kegunaan'] = nama_virus
```

```python
df
```

```python
df = df[~df['nama_obat'].isin(database['nama_obat'])]
```

```python
#df
```

```python
nama_obat = []
smiles = []
for obat in df['nama_obat']:
    print(obat)
    for compound in get_compounds(obat, 'name'):
        print(compound.cid)
        print(compound.canonical_smiles)
        nama_obat.append(compound.cid)
        smiles.append(compound.canonical_smiles)
```

```python
data_final = pd.DataFrame(nama_obat)
data_final['smiles'] = smiles
data_final = pd.DataFrame(data_final).rename(columns = {0 : 'cid'})
```

```python
data_final
```

```python
nama_virus = []
for i in data_final['cid']:
    nama_virus.append('antibiotics')
data_final['kegunaan'] = nama_virus
```

```python
data_final.to_csv('/antibiotics/antibiotics full.csv', index = False)
```
